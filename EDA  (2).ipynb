### 4.

```python
df['Wage'] = df['Wage'].str.replace('â‚¬', '')
df['Wage'] = df['Wage'].str.replace('K', '000')
df['Wage'] = df['Wage'].str.replace('M', '000000')
df['Wage'] = df['Wage'].astype(float)

# Plot histogram
plt.hist(df['Wage'])
plt.xlabel("Salary")
plt.ylabel("Number of Players")
plt.title("Salary Range of FIFA Players")
plt.show()
```

### 5.

```python
def height_to_cm(height):
    if isinstance(height, str):
        feet = int(height.split("'")[0])
        inches = int(height.split("'")[1].replace('"', ''))
        return feet * 30.48 + inches * 2.54
    else:
        return None

df['Height_cm'] = df['Height'].apply(height_to_cm)
```

```python
df.loc[df['Height_cm'].idxmax(), ['Name', 'Height']]
```
